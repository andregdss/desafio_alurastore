#1. Análise do faturamento



### Faturamento - Sem Frete:

```python
faturamento = {}
lojas = [loja1, loja2, loja3, loja4]

# Gerando os faturamentos:
for i, loja in enumerate(lojas, start=1):
    faturamento[f'Loja {i}'] = loja['Preço'].sum()

# Imprimindo os faturamentos:
print("O faturamento das lojas (sem frete) foram:")
for loja, valor in faturamento.items():
    print(f"{loja}: R$ {valor:.2f}\n")
```

```python
# Criando um DataFrame com esses faturamentos:
df_faturamento = pd.DataFrame(
    list(faturamento.items()),
    columns=['Loja', 'Faturamento']
)

df_faturamento
```

```python
# Conclusão sobre o faturamento:
idx_max = df_faturamento['Faturamento'].idxmax()
loja_max, valor_max = df_faturamento.loc[idx_max, ['Loja', 'Faturamento']]

print(f"A {loja_max} teve o maior faturamento de R${valor_max:.2f}.")
```

#### Gráfico:

```python
# Esquema de cores:
idx_max = df_faturamento['Faturamento'].idxmax()

cores = ['mediumorchid'] * len(df_faturamento)
cores[idx_max] = 'purple'

# Definindo tamanho da figura:
graf_faturamento = plt.figure(figsize=(8, 6))

# Criando o gráfico:
plt.bar(df_faturamento['Loja'], df_faturamento['Faturamento'], color=cores)

# Ajustes gerais:
plt.title('Comparativo de Faturamento das Lojas', fontsize=14, fontweight='bold')
plt.xlabel('')
plt.ylabel('')
plt.xticks(rotation=0, fontsize=12)
plt.yticks([])
plt.gca().set_facecolor('gainsboro')
plt.gca().spines[['top', 'left', 'right']].set_visible(False)

# Mostrar valores em cima das barras:
for i, valor in enumerate(df_faturamento['Faturamento']):
    plt.text(i, valor, f'R${valor:,.2f}', ha='center', va='bottom')

plt.show()
```

```python
graf_faturamento.savefig('graf_faturamento.png', bbox_inches='tight', dpi=300)
```

### Faturamento Com Frete:

```python
faturamento_com_frete = {}
lojas = [loja1, loja2, loja3, loja4]

# Gerando os faturamentos:
for i, loja in enumerate(lojas, start=1):
    faturamento_com_frete[f'Loja {i}'] = (loja['Preço'] + loja['Frete']).sum()

# Imprimindo os faturamentos:
print("O faturamento das lojas (com frete) foram:")
for loja, valor in faturamento_com_frete.items():
    print(f"{loja}: R$ {valor:.2f}\n")
```

```python
# Criando um DataFrame com esses faturamentos (com frete):
df_faturamento_com_frete = pd.DataFrame(
    list(faturamento_com_frete.items()),
    columns=['Loja', 'Faturamento']
)

df_faturamento_com_frete
```

```python
# Conclusão sobre o faturamento (com frete):
idx_max = df_faturamento_com_frete['Faturamento'].idxmax()
loja_max, valor_max = df_faturamento_com_frete.loc[idx_max, ['Loja', 'Faturamento']]

print(f"A {loja_max} teve o maior faturamento com frete de R${valor_max:.2f}.")
```

#### Gráfico:

```python
# Esquema de cores:
idx_max = df_faturamento_com_frete['Faturamento'].idxmax()

cores = ['greenyellow'] * len(df_faturamento_com_frete)
cores[idx_max] = 'lawngreen'

# Definindo tamanho da figura:
graf_faturamento_frete = plt.figure(figsize=(8, 6))

# Criando o gráfico:
plt.bar(df_faturamento_com_frete['Loja'], df_faturamento_com_frete['Faturamento'], color=cores)

# Ajustes gerais:
plt.title('Comparativo de Faturamento (com Frete) das Lojas', fontsize=14, fontweight='bold')
plt.xlabel('')
plt.ylabel('')
plt.xticks(rotation=0, fontsize=12)
plt.yticks([])
plt.gca().set_facecolor('gainsboro')
plt.gca().spines[['top', 'left', 'right']].set_visible(False)

# Mostrar valores em cima das barras:
for i, valor in enumerate(df_faturamento['Faturamento']):
    plt.text(i, valor + 1e5, f'R${valor:,.2f}', ha='center', va='bottom')

plt.show()
```

# 2. Vendas por Categoria

### Vendas por Categoria de cada Loja:

```python
# Gerando os DataFrames do total de vendas por categoria de cada Loja:
lojas = {
    'Loja 1': loja1,
    'Loja 2': loja2,
    'Loja 3': loja3,
    'Loja 4': loja4
}

vendas_categoria = {}

for loja, df in lojas.items():
    vendas_categoria[loja] = (df.groupby('Categoria do Produto').size().sort_values(ascending=False).reset_index(name='Total de Vendas'))
```

```python
# Exemplo de como chamar o DataFrame de cada loja:
vendas_categoria['Loja 1']
```

### Comparativo Geral de Vendas por Categoria:

```python
# Gerando DataFrame comparativo do total de vendas por categoria:

vendas_categorias={}

vendas_categorias['Loja 1'] = loja1['Categoria do Produto'].value_counts()
vendas_categorias['Loja 2'] = loja2['Categoria do Produto'].value_counts()
vendas_categorias['Loja 3'] = loja3['Categoria do Produto'].value_counts()
vendas_categorias['Loja 4'] = loja4['Categoria do Produto'].value_counts()

df_vendas_categorias = pd.DataFrame(vendas_categorias)

df_vendas_categorias['Maior Venda'] = df_vendas_categorias.idxmax(axis=1)

df_vendas_categorias
```

```python
df_vendas_categorias['Maior Venda']
```

### Conclusão:

```python
# Conclusão sobre Vendas por Categoria:

for loja, df in vendas_categoria.items():
    idx = df['Total de Vendas'].idxmax()
    categoria = df.loc[idx, 'Categoria do Produto']
    total = df.loc[idx, 'Total de Vendas']

    print(f'A categoria mais vendida na {loja} é "{categoria}", com {total} vendas.\n')
```

### Gráfico:

```python
# Ajustando o DataFrame
df = df_vendas_categorias.drop(columns=['Maior Venda'])

# Criando o gráfico de barras empilhadas:
df.plot(kind='bar', stacked=True, figsize=(12, 8), colormap='Blues')

# Ajustes no gráfico:
plt.title('Vendas por Categoria nas Lojas', fontsize=14, fontweight='bold')
plt.xlabel('')
plt.ylabel('')
plt.xticks(rotation=45)
plt.yticks(range(0, 2001, 100))
plt.legend(title='Categoria de Produto')
plt.gca().set_facecolor('gainsboro')
plt.gca().spines[['top', 'right']].set_visible(False)

plt.grid(axis='y', alpha=0.5)
plt.savefig('graf_vendas_categoria.png', bbox_inches='tight', dpi=300)
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

### Considerando os 5 mais vendidos e os 5 menos vendidos em cada loja:

```python
# Criando a função para exibir os 5 mais e os 5 menos vendidos:
def exibir_mais_menos_vendidos(df_loja, numero_loja):
    produtos = df_loja.groupby('Produto').size().sort_values(ascending=False).reset_index(name='Quantidade Vendida')
    mais_vendidos = produtos.head(5)
    menos_vendidos = produtos.tail(5)

    print(f'Os produtos mais vendidos na Loja {numero_loja} são:\n{mais_vendidos}\n')
    print(f'Os produtos menos vendidos na Loja {numero_loja} são:\n{menos_vendidos}\n')
```

```python
# Chamando a função para a Loja 1:
exibir_mais_menos_vendidos(loja1, 1)
```

```python
# Chamando a função para a Loja 2:
exibir_mais_menos_vendidos(loja2, 2)
```

```python
# Chamando a função para a Loja 3:
exibir_mais_menos_vendidos(loja3, 3)
```

```python
# Chamando a função para a Loja 4:
exibir_mais_menos_vendidos(loja4, 4)
```

```python
# Para usar posteriormente no gráfico:

# Função que retorna mais e menos vendidos:
lojas = [loja1, loja2, loja3, loja4]

def mais_menos_vendidos(df_loja):
    produtos = df_loja.groupby('Produto').size().sort_values(ascending=False).reset_index(name='Quantidade Vendida')
    return produtos.head(5), produtos.tail(5)

# Criando dicionários para armazenar os DataFrames de cada loja:
mais_vendidos = {}
menos_vendidos = {}

for i, loja in enumerate(lojas, start=1):
    mais, menos = mais_menos_vendidos(loja)
    mais_vendidos[f'Loja {i}'] = mais
    menos_vendidos[f'Loja {i}'] = menos

# Acessando os DataFrames:
df_mais_vendidos_loja1 = mais_vendidos['Loja 1']
df_menos_vendidos_loja1 = menos_vendidos['Loja 1']

df_mais_vendidos_loja2 = mais_vendidos['Loja 2']
df_menos_vendidos_loja2 = menos_vendidos['Loja 2']

df_mais_vendidos_loja3 = mais_vendidos['Loja 3']
df_menos_vendidos_loja3 = menos_vendidos['Loja 3']

df_mais_vendidos_loja4 = mais_vendidos['Loja 4']
df_menos_vendidos_loja4 = menos_vendidos['Loja 4']
```

```python
df_mais_vendidos_loja1
```

#### Gráfico:

```python
def grafico_mais_menos(mais_df, menos_df, nome_loja):
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))
    fig.suptitle(f"Produtos Mais e Menos Vendidos - {nome_loja}", fontsize=16, fontweight='bold')

    # Mais Vendidos:
    ax = axes[0]
    ax.barh(mais_df["Produto"], mais_df["Quantidade Vendida"], color='firebrick')
    ax.invert_yaxis()
    ax.set_title("Top 5 Mais Vendidos", fontweight='bold')
    ax.set_xlabel("Quantidade Vendida")

    ax.grid(axis='x', linestyle='--', alpha=0.5)
    ax.set_facecolor('gainsboro')
    ax.spines[['top', 'right']].set_visible(False)


    # Menos Vendidos:
    ax = axes[1]
    ax.barh(menos_df["Produto"], menos_df["Quantidade Vendida"], color='Salmon')
    ax.invert_yaxis()
    ax.set_title("Top 5 Menos Vendidos", fontweight='bold')
    ax.set_xlabel("Quantidade Vendida")

    ax.grid(axis='x', linestyle='--', alpha=0.5)
    ax.set_facecolor('gainsboro')
    ax.spines[['top', 'right']].set_visible(False)

    # Exibir:
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    fig.savefig(f"graf_mais_menos_{nome_loja}.png", dpi=300)
    plt.show()
```

```python
import matplotlib.cm as cm

def grafico_pizza(mais_df, menos_df, nome_loja):

    fig, axes = plt.subplots(1, 2, figsize=(14, 6))
    fig.suptitle(f"Produtos Mais e Menos Vendidos - {nome_loja}", fontsize=16, fontweight='bold')
    cores = cm.Reds(np.linspace(0, 1, len(mais_df)))

    # Gráfico Esquerda: Mais Vendidos
    axes[0].pie(
        mais_df["Quantidade Vendida"],
        labels=mais_df["Produto"],
        autopct="%1.1f%%",
        startangle=90,
        textprops={'fontsize': 12},
        colors=cores,
        wedgeprops={'edgecolor': 'silver', 'linewidth': 1}
    )
    axes[0].set_title(f"Top 5 – Mais Vendidos", fontsize=14, fontweight='bold')

    # Gráfico Direita: Menos Vendidos
    axes[1].pie(
        menos_df["Quantidade Vendida"],
        labels=menos_df["Produto"],
        autopct="%1.1f%%",
        startangle=90,
        textprops={'fontsize': 12},
        colors=cores,
        wedgeprops={'edgecolor': 'silver', 'linewidth': 1}
    )
    axes[1].set_title(f"Top 5 – Menos Vendidos", fontsize=14, fontweight='bold')

    plt.tight_layout()
    fig.savefig(f"graf_pizza_{nome_loja}.png", dpi=300)
    plt.show()
```

# 5. Frete Médio por Loja

```python
# Montando um DataFrame comparativo:

lojas = [loja1, loja2, loja3, loja4]

df_frete_medio = pd.DataFrame({
    'Lojas': [f'Loja {i}' for i in range(1, 5)],
    'Frete Médio': [loja['Frete'].mean().round(2) for loja in lojas]
})

df_frete_medio
```

```python
# Conclusão da Loja com maior valor de Frete Médio:

idx_max = df_frete_medio['Frete Médio'].idxmax()
loja_max, valor_max = df_frete_medio.loc[idx_max, ['Lojas', 'Frete Médio']]

print(f'A {loja_max} tem o maior valor de frete de R${valor_max}.')
```

```python
# Conclusão da Loja com menor valor de Frete Médio:

idx_min = df_frete_medio['Frete Médio'].idxmin()
loja_min, valor_min = df_frete_medio.loc[idx_min, ['Lojas', 'Frete Médio']]

print(f'A {loja_min} tem o maior valor de frete de R${valor_min}.')
```

#### Gráficos:

```python
cores = ['mediumvioletred', 'deeppink', 'hotpink', 'palevioletred']

graf_frete_medio = plt.figure(figsize=(10,6))
plt.barh(df_frete_medio['Lojas'], df_frete_medio['Frete Médio'], color=cores)


# Títulos
plt.title("Frete Médio por Loja", fontsize=16, fontweight='bold')
plt.xlabel("Valor (R$)")

# Exibir valores no final das barras:
for i, v in enumerate(df_frete_medio['Frete Médio']):
    plt.text(v + 0.05, i, f"R$ {v}", va='center', fontweight='bold')

# Ajustes visuais:
plt.gca().spines[['top', 'right']].set_visible(False)
plt.gca().set_facecolor('gainsboro')
plt.grid(axis='x', linestyle='--', alpha=0.5)

# Exibindo:
plt.tight_layout()
plt.show()
```
